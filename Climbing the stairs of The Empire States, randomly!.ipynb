# A Hacker Statistics Approach To A Simple Game


Walking up the empire state building, you throw a dice 1000 times, if it's 1 or 2 you'll one step down.
if it's 3, 4, 5 you2ll go one step up. If you throw a 6, you roll the die again and you will walk up the resulting number of steps. you can't go lower than step number 0. Also you're a bit clumsy and you have %1 chance of falling down the stairs where you start. You bet with your friend that you'll reach 60 steps high.
What is the chance that you'll win this bet?

We're going to find out by simulating the scenario here in this notebook.

First we start with the basic element of the game. Rolling a die. For that matter, we first import numpy and use it's random module to simulate rolling a die.

```python
import numpy as np
np.random.seed(123)
print(np.random.randint(1,7))

```

In the Empire State Building bet, your next move depends on the number of eyes you throw with the dice. We can perfectly code this with an if-elif-else construct!

```python
# Numpy is imported, seed is set

# Starting step
step = 50

# Roll the dice
dice = np.random.randint(1,7)


# Finish the control construct
if dice <= 2 :
    step = step - 1
elif dice < 6 :
    step = step + 1
else :
    step = step + np.random.randint(1,7)

# Print out dice and step
print("Dice was: " + str(dice))
print("Step is now: " + str(step))
```

Now with this process we have the most basic element handed down for our game: The Random Walk!

Whether we are going to step up or down is handled by the randomness of a dice being rolled. What we're going to need right now is to reproduce this approach for 1000 times! Plus, we're going to have to record our steps. In order to do that we're gonna save them, a list is going to be our approach here.

```python
# Numpy is imported, seed is set

# Initialize random_walk
random_walk = [0]

# Complete the ___
for x in range(100) :
    # Set step: last element in random_walk
    step = random_walk[-1]

    # Roll the dice
    dice = np.random.randint(1,7)

    # Determine next step
    if dice <= 2:
        step = step - 1
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    # append next_step to random_walk
    random_walk.append(step)

# Print random_walk
print(random_walk)
```

Great! Our random walk is working. But there are negative values as our step count. Which shouldn't be possible since there are no below ground stairs. We're going to solve the problem by using the max function on our if statement that sets the step count to a negative value.

```python
# Numpy is imported, seed is set

# Initialize random_walk
random_walk = [0]

for x in range(100) :
    step = random_walk[-1]
    dice = np.random.randint(1,7)

    if dice <= 2:
        # Replace below: use max to make sure step can't go below 0
        step = max(0, step - 1)
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    
    random_walk.append(step)

print(random_walk)
```

Great. Now that is fixed, we could take a look broader look at our steps by visualizing our list of random walk.

```python
# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Plot random_walk
plt.plot(random_walk)

# Show the plot
plt.show()
```

Now we're read to learn about the distribution of random walks.

The initial problem was you throw a die a thousand times, depending on the result you go a step up or down. What is the chance that you'll reach 60 steps high?

Since each random walk is going to end up on different steps, what we're going to want here is to simulation of these random walks for thousands of times.


```python
# Numpy is imported; seed is set

# Initialize all_walks (don't change this line)
all_walks = []

# Simulate random walk 10 times
for i in range(10):

    # Code from before
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)

    # Append random_walk to all_walks
    all_walks.append(random_walk)

# Print all_walks
print(all_walks)
```

Now that we have all of our random walks recorded. We can visualize them!

```python
# Convert all_walks to Numpy array: np_aw
np_aw = np.array(all_walks)

# Plot np_aw and show
plt.plot(np_aw)
plt.show()

# Clear the figure
plt.clf()

# Transpose np_aw: np_aw_t
np_aw_t = np.transpose(np_aw)

# Plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()
```

Without transposing the numpy array, Python misunderstands our data.

There's still something we forgot. The last rule we need to implement is the on the off chance that we fall from stairs, we're going to have to go back to ground level, and start from zero.

```python
# numpy and matplotlib imported, seed set

# Simulate random walk 250 times
all_walks = []
for i in range(250) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)

        # Implement clumsiness
        if np.random.rand() <= 0.001 :
            step = 0

        random_walk.append(step)
    all_walks.append(random_walk)

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))
plt.plot(np_aw_t)
plt.show()
```

All these fancy visualizations have put us on a sidetrack. We still have to solve the million-dollar problem: What are the odds that you'll reach 60 steps high on the Empire State Building?

Basically, you want to know about the end points of all the random walks you've simulated. These end points have a certain distribution that you can visualize with a histogram.

```python
# numpy and matplotlib imported, seed is set

# Simulate random walk 500 times
np.random.seed(123)
all_walks = []
for i in range(500) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        if np.random.rand() <= 0.001 :
            step = 0
        random_walk.append(step)
    all_walks.append(random_walk)

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))

# Select last row from np_aw_t: ends
ends = np_aw_t[-1,:]

# Plot histogram of ends, display plot
plt.hist(ends)
plt.show()
```

So what are the chances?
We have the array `ends` containing the conclusions, the last steps we ended up with in our every single trials. So what we're going to need is to find out how many of them ended up at 60 or above and divide that number by 500, which is the length of all trials. 

```python
len(ends[ends >= 60]) / 500
```

Seems like you have a pretty high chance of winning the bet!
